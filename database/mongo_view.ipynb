```python
from pymongo import MongoClient
from dotenv import load_dotenv
from urllib.parse import quote_plus
import os
```

```python
load_dotenv(override=True)
```

```python
user = os.getenv('MONGO_USER', '')
password = os.getenv('MONGO_PASSWORD', '')
host = os.getenv('MONGO_HOST', '')
port = os.getenv('MONGO_PORT', '')

db_name = os.getenv('DATABASE', '')

uri = f"mongodb://{user}:{password}@{host}:{port}/"
```

```python
with MongoClient(uri) as client:  # Remove underscores
    collection = client[db_name]['carmichael_number']
    # collection.drop()
    # result = collection.aggregate([
    #     {"$group": {
    #         "_id": None,
    #         "max_id": {"$max": "$_id"}
    #     }}
    # ]).next()

    # print(result["max_id"])
    result = collection.aggregate([
        {"$limit": 1},
        {"$project": {
            "_id": 1,
            "_id_type": {"$type": "$_id"},
            "factors": 1,
            "factors_type": {"$type": "$factors"},
            "first_factor": {"$arrayElemAt": ["$factors", 0]},
            "first_factor_type": {"$type": {"$arrayElemAt": ["$factors", 0]}}
        }}
    ]).next()
    print(result)
```
